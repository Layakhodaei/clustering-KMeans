```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
from sklearn.cluster import KMeans
```

```python
# Read Images
img = mpimg.imread('E:\\AUT\\Machine Learning\\PML_HW3\\DataSet\\3\\flower.jpg')
 
# Output Images
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(img);
```

```python
img.shape
```

```python
data = img / 255.0 # use 0...1 scale
data = data.reshape(194 * 259, 3)
data.shape
```

```python
import warnings; warnings.simplefilter('ignore')  # Fix NumPy issues.
#K is 2,3,5,8,10,15,30
kmeans = KMeans(30)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]
```

```python
img_recolored = new_colors.reshape(img.shape)

fig, ax = plt.subplots(1, 2, figsize=(12, 6), subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(img)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(img_recolored)
ax[1].set_title('30-color Image', size=16);
```
